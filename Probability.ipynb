### Basic probability concepts

Sure! Probability is a branch of mathematics that deals with the likelihood of an event occurring. Here are some basic concepts and terms related to probability:

1. **Experiment:** An action or process that leads to one of several possible outcomes. For example, tossing a coin is an experiment with two possible outcomes: heads or tails.

2. **Sample Space (S):** The set of all possible outcomes of an experiment. For a single coin toss, the sample space is {H, T}, where H represents heads and T represents tails.

3. **Event (E):** A subset of the sample space. An event can be a single outcome or a combination of outcomes. Using the coin toss example, getting heads can be an event.

4. **Probability (P):** A measure that quantifies the likelihood of an event occurring. It's a number between 0 and 1, inclusive.
   - 0≤P(E)≤1

5. **Probability of an Event:** The probability of an event E occurring is calculated as:
   P(E)= Total number of Number of favorable outcomes / possible outcomes
​


6. **Complementary Event:** The event that an event E does not occur, denoted by \( E' \) or \( \neg E \). The probability of the complementary event is:
    P(E') = 1 - P(E) 

7. **Mutually Exclusive Events:** Two events that cannot occur at the same time. If event A and event B are mutually exclusive, then:
   P(A or B) = P(A) + P(B) 

8. **Independent Events:** Two events are independent if the occurrence of one does not affect the occurrence of the other. The probability of two independent events A and B both occurring is:
   P(A and B) = P(A) X P(B) 

9. **Conditional Probability:** The probability of an event occurring given that another event has already occurred. It's denoted by  P(A|B)  and calculated as:
   P(A|B) = P(A  and B)/P(B)

10. **Bayes' Theorem:** A formula that describes the probability of an event based on prior knowledge of conditions that might be related to the event. It's expressed as:
    P(A|B) = P(A)  X P(B)/P(B)

These are some foundational concepts in probability. Understanding these can help you solve various problems involving uncertainty and randomness.

Probability distributions describe how the probabilities of different outcomes are distributed across the values of a random variable. There are two main types of probability distributions: discrete and continuous.

### Discrete Probability Distribution

In a discrete probability distribution, the random variable can take on a countable number of distinct values. The probability distribution is defined by a probability mass function (PMF) which gives the probability of each possible value of the discrete random variable.

**Characteristics of Discrete Probability Distribution:**
- Values are countable and distinct.
- The sum of all probabilities equals 1.
- The probability of each specific value is between 0 and 1.

**Examples:**
- **Bernoulli Distribution:** Describes a single experiment with two possible outcomes (e.g., success or failure).
- **Binomial Distribution:** Describes the number of successes in a fixed number of independent Bernoulli trials.
- **Poisson Distribution:** Describes the number of events occurring in a fixed interval of time or space.

### Continuous Probability Distribution

In a continuous probability distribution, the random variable can take on an infinite number of possible values within a given range. The probability distribution is defined by a probability density function (PDF) which gives the probability density (or likelihood) of the random variable taking on a particular value.

**Characteristics of Continuous Probability Distribution:**
- Values are uncountably infinite and can take any value within a range.
- The total area under the PDF curve equals 1.
- Probability for a specific value is typically 0 (since there are infinitely many possible values).

**Examples:**
- **Uniform Distribution:** Describes a continuous random variable where all intervals of the same length have an equal probability.
- **Normal (Gaussian) Distribution:** Describes data that clusters around a mean with the data points symmetrically distributed around the mean.
- **Exponential Distribution:** Describes the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate.

### Key Differences:

1. **Nature of Values:**
   - **Discrete:** Countable and distinct values.
   - **Continuous:** Infinite and uncountable values within a range.

2. **Representation:**
   - **Discrete:** Probability Mass Function (PMF)
   - **Continuous:** Probability Density Function (PDF)

3. **Probability Calculation:**
   - **Discrete:** Individual probabilities for each distinct value.
   - **Continuous:** Probability as area under the PDF curve over an interval.

Understanding these types of probability distributions is fundamental in statistics and data analysis. They help in modeling real-world phenomena, making predictions, and understanding the behavior of random variables in various scenarios.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0
sigma = 1

# Range of values
x = np.linspace(-5, 5, 1000)

# PDF formula for normal distribution
pdf = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Normal Distribution PDF', color='blue')
plt.title('Probability Density Function (PDF) - Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.stats import binom

# Parameters
n = 10
p = 0.5

# Range of values
k = np.arange(0, n+1)

# PMF formula for binomial distribution
pmf = binom.pmf(k, n, p)

# Plot
plt.figure(figsize=(8, 6))
plt.stem(k, pmf, label='Binomial Distribution PMF', linefmt='blue', markerfmt='bo', basefmt=" ")
plt.title('Probability Mass Function (PMF) - Binomial Distribution')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

# PDF of standard normal distribution
def pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)

# Integrate PDF from -infinity to +infinity
integral, _ = spi.quad(pdf, -np.inf, np.inf)

print("Total area under the PDF curve:", integral)

# Plot PDF
x = np.linspace(-4, 4, 1000)
y = pdf(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Standard Normal Distribution PDF', color='blue')
plt.fill_between(x, y, 0, alpha=0.3, color='blue')
plt.title('Standard Normal Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
